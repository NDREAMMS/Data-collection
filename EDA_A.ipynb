# Feuille de route EDA — ImmoVision 360 (Périmètre Élysée)

## Question 1
**Hypothèse** : Peu d'hôtes concentrent une grande part de l'offre.  
**Variable(s)** : `calculated_host_listings_count` (quantitative — proxy de concentration)  
**Graphique** : Histogramme + courbe de fréquences cumulées. Axe x = nombre d'annonces par hôte, axe y = fréquence. Filtre : exclure les −1.

## Question 2
**Hypothèse** : Les logements très disponibles ont davantage un profil "catalogue".  
**Variable(s)** : `availability_365` (quantitative) × `standardization_score` (score −1/0/1)  
**Graphique** : Boîtes à moustaches de `availability_365` par valeur de `standardization_score`. Traitement à part des −1.

## Question 3
**Hypothèse** : Les hôtes très réactifs gèrent plusieurs logements (profil professionnel).  
**Variable(s)** : `host_response_rate_num` (quantitative) × `calculated_host_listings_count` (quantitative)  
**Graphique** : Nuage de points. Axe x = taux de réponse, axe y = nombre d'annonces. Couleur = `room_type_code`.

## Question 4
**Hypothèse** : Le type de logement influence l'impact sur le quartier.  
**Variable(s)** : `room_type_code` (nominale) × `neighborhood_impact_score` (score −1/0/1)  
**Graphique** : Barres empilées en % par type de logement. Exclure ou isoler les −1.

## Question 5
**Hypothèse** : La réactivité varie selon le type de logement.  
**Variable(s)** : `host_response_time_code` (ordinale) × `room_type_code` (nominale)  
**Graphique** : Tableau croisé (heatmap ou barres groupées). Axe x = type de logement, axe y = code de délai.
